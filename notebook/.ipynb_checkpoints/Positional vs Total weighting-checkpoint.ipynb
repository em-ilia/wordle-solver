```python
import matplotlib.pyplot as plt
import numpy as np
from sys import path
path.insert(0, '..')
from game import GameManager
```

```python
x = np.arange(0,1,0.01)
```

```python
Gs = []
for i in x:
    Gs.append(GameManager((i,1-i), count=100))
```

```python
for g in Gs:
    g.run()
```

```python
y = [g.compute_avgAttempts() for g in Gs]
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x, y)
plt.ylabel('Average attempts required')
plt.xlabel('Positional weighting')
plt.title('Avg. Attempts vs Positional Weighting\n(lower is better)')
plt.gcf().set_facecolor('#ee9999')
plt.savefig('graph1.png', bbox_inches='tight', pad_inches=0, dpi=300)
```
